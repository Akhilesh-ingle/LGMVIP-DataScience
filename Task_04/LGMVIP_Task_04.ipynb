## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

## Loading Image

```python
image = cv2.imread("car3.jpg")
#plt.imshow(image)
display(Image.fromarray(image))
```

## Converting into RGB Image

```python
original_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(original_image))
```

## Converting into Grayscale Image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#plt.imshow(gray_image)
display(Image.fromarray(gray_image))
```

## Inverting The Grayscale Image

```python
inverted_image = 255 - gray_image
#plt.imshow(inverted_image)
display(Image.fromarray(inverted_image))
```

## Converting into Blurry Image

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
#plt.imshow(blurred)
display(Image.fromarray(blurred))
```

## Inverting the Blurry Image

```python
inverted_blurred = 255 - blurred
#plt.imshow(pencil_sketch)
display(Image.fromarray(inverted_blurred))
```

## Making Pencil Sketch of Image

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
display(Image.fromarray(pencil_sketch))
```
